# ST Louis Blues
All data comes from Hockeydb sites below.

[St. Louis Blues](http://www.hockeydb.com/stte/st.-louis-blues-8187.html)
 & 
[Boston Bruins](http://www.hockeydb.com/stte/boston-bruins-4919.html)

Downloaded both sites to file: stlb_site.html & bb_site.html, respectively.

```python
import pandas as pd
```

```python
stlb_df = pd.read_html('stlb_site.html')[0]
bb_df = pd.read_html('bb_site.html')[0]
```

```python
stlb_df.to_csv('NHL_STL_Blues_2019.csv', index=False)
bb_df.to_csv('NHL_Boston_Bruins_2019.csv', index=False)
```
